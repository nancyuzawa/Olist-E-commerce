# Seção 1 - Introdução e Motivação


O grupo trabalhará na **previsão de demandas de volume de vendas e faturamento por categoria de produto com base no histórico pedidos do e-commerce**.

Diante disso, para compreender melhor os padrões das vendas, pretendemos explorar/analisar as seguintes questões:

* Quais fatores mais influenciam o valor das vendas?

# Seção 2 - Descrição da Base de Dados


A base de dados foi obtida através do site da Kaggle, no link: https://www.kaggle.com/datasets/olistbr/brazilian-ecommerce.

Esses dados são reais que mostram os pedidos realizados entre 2016 e 2018 na plataforma Olist.

O dataset Olist é composto por 9 arquivos no formato CSV, em que cada arquivo representa uma tabela distinta. Essas tabelas são conectadas por IDs, formando uma base de dados relacional. Assim, cada linha possui um significado variando de acordo com a tabela analisada.


* **olist_customers_dataset**: a linha contém as informações sobre o cliente e a sua localização
* **olist_geolocation_dataset**: a linha contém informações sobre os CEPs brasileiros e duas coordenadas(latitude e longitude)
* **olist_order_items_dataset**: a linha contém as informações sobre os itens comprados em cada pedido com dados como o vendedor, o produto, preço e valor do frete.
* **olist_order_payments_dataset**: informações sobre as opções de pagamento dos pedidos com o ID do pedido, tipo de pagamento, número de parcelas e valor.
* **olist_order_reviews_dataset**: dados de avaliações do cliente, como a pontuação, título do comentário, mensagem da avaliação e data e hora da publicação.
* **olist_orders_dataset**: contém dados do pedido, sendo o principal, com dados do status do pedido, a data e hora da compra/aprovação de pagamento, data de entrega e data de entrega que foi estimada.
* **olist_products_dataset**: informações sobre o produto vendido, como a categoria, quantidade de fotos, o peso em gramas do produto, comprimento, altura, largura em centímetros.
* **olist_sellers_dataset**: informações sobre os vendedores que processaram o pedido, como os cinco primeiros dígitos do CEP, a cidade e o estado do vendedor.
* **product_category_name_translation**: a linha mostra o nome da categoria em português e em inglês.

### Principais atributos utilizados na análise

Foi feita a união com merge das tabelas de dados(orders, products, customers, review).

De forma geral, a base não apresentou problemas críticos de organização, porém exigiu etapas importantes de tratamento, principalmente relacionadas a tipos de dados, padronização e validação das informações após o cruzamento das tabelas.

Um dos primeiros pontos verificados foi a existência de registros duplicados, utilizando o método duplicated(). Essa etapa é essencial após operações de junção (merge), já que relacionamentos do tipo um-para-muitos podem gerar duplicidade de linhas. A verificação permitiu validar se o processo de integração das bases manteve a integridade dos dados.

Em seguida, foi realizada a correção dos tipos de dados, especialmente para colunas de datas que estavam no formato de texto (object). Essas colunas foram convertidas para o tipo datetime, o que é fundamental para permitir análises temporais mais precisas. Posteriormente, a coluna de data de compra (order_purchase_timestamp) foi ajustada para conter apenas a data (sem horário), facilitando agregações por dia.

Também foi feito um pequeno ajuste de padronização de nomenclatura, com a renomeação da coluna order_item_id para qtd_item, tornando seu significado mais claro dentro do contexto da análise.

Ao longo da exploração, foram realizadas diversas agregações e agrupamentos para entender o comportamento dos dados, como:

volume de vendas por categoria de produto,
faturamento por categoria,
distribuição de vendas por estado e cidade,
desempenho das categorias com base nas avaliações dos clientes.

Essas análises ajudaram a identificar padrões e preparar os dados para etapas posteriores, como modelagem.

Na construção da base de treinamento, foi feita uma seleção das variáveis mais relevantes, com a remoção de colunas que não agregariam valor ao modelo, como identificadores (order_id, customer_id, etc.) e campos temporais intermediários que não seriam utilizados diretamente.

Sobre dados faltantes (valores nulos), foram realizadas verificações com isna().sum(). Como estratégia de tratamento, optou-se pela remoção das linhas com valores nulos (dropna()), garantindo que o modelo fosse treinado apenas com dados completos. Essa abordagem é adequada quando a quantidade de valores ausentes não compromete significativamente o volume de dados.

## 3.1 – Importações e Carregamento dos Dados

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = {
 'items' : pd.read_csv('./olist_order_items_dataset.csv', encoding='latin1'),
 'orders' : pd.read_csv('./olist_orders_dataset.csv', encoding='latin1'),
 'products' : pd.read_csv('./olist_products_dataset.csv', encoding='latin1'),
 'customers': pd.read_csv('./olist_customers_dataset.csv', encoding='latin1'),
 'review' : pd.read_csv('./olist_order_reviews_dataset.csv', encoding='latin1'),
}
```

## 3.2 – Junção das Tabelas

```python
# Une as tabelas
merged = df['orders'].merge(df['items'], on='order_id', how='inner')
merged = merged.merge(df['products'], on='product_id', how='inner')
merged = merged.merge(df['customers'], on='customer_id', how='inner')
merged = merged.merge(df['review'], on='order_id', how='left')

merged.head()
```

## 3.3 – Limpeza de dados

Antes de qualquer transformação, verificamos a existência de registros duplicados e de valores ausentes (nulos) em todas as colunas do DataFrame consolidado:

```python
# Removendo duplicados
clean = merged.drop_duplicates()

# Removendo valores nulos 
clean.dropna(inplace=True)
```

```python
print(clean.isna().sum())
```

## 3.4 – Conversão de Tipos e Renomeação de Colunas

```python
# Convertendo colunas de data de object para datetime64
date_cols = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 
             'order_delivered_customer_date', 'order_estimated_delivery_date', 'shipping_limit_date']

for col in date_cols:
    clean[col] = pd.to_datetime(clean[col])
```

```python
# Renomeando a coluna order_item_id
clean.rename(columns={'order_item_id': 'qtd_item'}, inplace=True)
```

## 3.5 - Pré-processamento de Atributos

```python
# Garantindo tipo datetime
clean['order_purchase_timestamp'] = pd.to_datetime(clean['order_purchase_timestamp'])

# Criando variáveis de tempo (muito importante para análise)
clean['ano'] = clean['order_purchase_timestamp'].dt.year
clean['mes'] = clean['order_purchase_timestamp'].dt.month
clean['dia'] = clean['order_purchase_timestamp'].dt.day
clean['dia_semana'] = clean['order_purchase_timestamp'].dt.weekday

# (Opcional mas recomendado) tratar outliers de preço
clean = clean[clean['price'] < clean['price'].quantile(0.99)]
```

## 3.6 – Agrupamentos para Análise Exploratória

```python
# DataFrame completo com todas as dimensões de análise
order_full = clean.groupby([
    'order_id', 'product_category_name', 'customer_id',
    'order_status', 'order_purchase_timestamp', 'order_approved_at',
    'order_delivered_carrier_date', 'order_delivered_customer_date',
    'order_estimated_delivery_date', 'product_id', 'seller_id',
    'shipping_limit_date', 'price', 'freight_value',
    'customer_unique_id', 'customer_city', 'customer_state'
])['qtd_item'].count().reset_index()
order_full
```

# Seção 4 - Questões de exploração (EDA)

**1. Quais categorias de produtos possuem maior volume de vendas?**

Hipótese: Algumas categorias concentram a maior parte das vendas.

Análise: Foi realizado um agrupamento por product_category_name, considerando a quantidade de itens vendidos (qtd_item).

A hipótese foi validada: o volume de vendas está concentrado em poucas categorias, especialmente cama_mesa_banho e beleza_saude.

```python
products = clean.groupby('product_category_name')['qtd_item'].count().reset_index()
```

```python
import matplotlib.pyplot as plt

# Ordenando e pegando Top 10
top_products = products.sort_values(by='qtd_item', ascending=False).head(10)

# Criando o gráfico
plt.figure(figsize=(12,6))
plt.bar(top_products['product_category_name'], top_products['qtd_item'])

# Ajustes visuais
plt.title('Top 10 Categorias por Volume de Vendas')
plt.xlabel('Categoria do Produto')
plt.ylabel('Quantidade de Itens Vendidos')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

Esse gráfico é do tipo barra (bar chart). Ele é adequado, pois possibilita comparar entre as categorias e possibilita visuaizar as Top 10 categorias por volume de vendas (qtd_item).

**2. Quais categorias geram maior faturamento?**

 Hipótese: As categorias mais vendidas não são necessariamente as mais lucrativas.

 Análise: Soma do price por categoria.

 Observa-se que algumas categorias, mesmo não liderando em volume, apresentam maior faturamento, indicando maior valor agregado por produto. Como vemos no gráfico abaixo os maiores valores de receita de venda derivam, das categorias 'beleza_saude' e 'relogios_presentes', somente em terceiro aparecendo o setor com o maior numero de vendas 'cama_mesa_banho'.

```python
products_price = clean.groupby(['product_category_name']).agg({"price":"sum", "qtd_item":"count"})
```

```python
import matplotlib.pyplot as plt

# Ordenando e pegando Top 10
top_faturamento = products_price.sort_values(by='price', ascending=False).head(10).reset_index()

# Criando o gráfico
plt.figure(figsize=(12,6))
plt.bar(top_faturamento['product_category_name'], top_faturamento['price'])

# Ajustes visuais
plt.title('Top 10 Categorias por Faturamento')
plt.xlabel('Categoria do Produto')
plt.ylabel('Faturamento Total (R$)')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

Também é do tipo barra (bar chart). Ele é adequado, pois permite a comparação dos valores e a classificação das categorias que obtiveram os 10 maiores faturamentos.

**3. Existe relação entre quantidade vendida e faturamento?**

Hipótese: Existe correlação positiva entre quantidade de itens e faturamento.

Análise:Comparação entre qtd_item e price.

Há uma correlação positiva, porém não linear forte, indicando que o preço médio influencia significativamente o faturamento.

```python
relacao = clean.groupby('product_category_name').agg({
    'qtd_item':'count',
    'price':'sum'
}).reset_index()

plt.figure(figsize=(8,6))
plt.scatter(relacao['qtd_item'], relacao['price'])

plt.title('Quantidade vs Faturamento por Categoria')
plt.xlabel('Quantidade de Itens')
plt.ylabel('Faturamento')
plt.tight_layout()
plt.show()
```

Aqui está utilizando o gráfico de dispersão (scatter plot). Ele é mais adequado para esta questão, pois permite ver há uma relação da quantidade de itens vendidos com o faturamento gerado, permitindo identificar tendências.

**4. Como as vendas estão distribuídas entre os estados?**

Hipótese: Estados mais populosos concentram mais vendas.

Análise: Agrupamento por customer_state.

A hipótese foi confirmada. O volume de vendas apresenta forte concentração em estados mais populosos e economicamente relevantes do país.

```python
vendas_estado = clean.groupby('customer_state')['qtd_item'].count() \
    .sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(10,5))
plt.bar(vendas_estado['customer_state'], vendas_estado['qtd_item'])

plt.title('Top 10 Estados por Volume de Vendas')
plt.xlabel('Estado')
plt.ylabel('Quantidade de Itens')
plt.tight_layout()
plt.show()
```

Aqui utilizamos gráfico de barras para comparar o volume de vendas (quantiade de itens) nos estados, permitindo visualizar os 10 primeiros estados que possuem a maior quantidade de vendas realizadas.

**5. Existe variação significativa nas vendas entre cidades dentro do mesmo estado?**

Hipótese: Mesmo dentro de um estado, há concentração em grandes centros urbanos.

Análise: Agrupamento por customer_state e customer_city.

A hipótese se confirma, pois observa-se variação significativa nas vendas entre cidades de um mesmo estado, caracterizada por forte concentração nos grandes centros urbanos.

```python
vendas_cidade = clean.groupby(['customer_state','customer_city'])['qtd_item'].count() \
    .sort_values(ascending=False).head(10).reset_index()

labels = vendas_cidade['customer_city'] + " - " + vendas_cidade['customer_state']

plt.figure(figsize=(12,6))
plt.barh(labels, vendas_cidade['qtd_item'])

plt.title('Top 10 Cidades por Volume de Vendas')
plt.xlabel('Quantidade')
plt.ylabel('Cidade')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

Aqui também utilizamos o gráfico de barras horizontais para comparar o volume de vendas (quantidade de itens) entre as cidades dentro dos estados, permitindo visualizar as 10 cidades com maior volume de vendas.

**6. Como os preços estão distribuídos?**

Hipótese: A distribuição de preços é assimétrica (skewed).

Análise: Distribuição da variável price.

A hipótese se confirma, pois a distribuição de preços apresenta forte assimetria à direita, com concentração em valores baixos e uma cauda longa de preços elevados.

```python
plt.figure(figsize=(8,6))
plt.hist(clean['price'], bins=50)

plt.title('Distribuição dos Preços')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.tight_layout()
plt.show()
```

Foi utilizado um histograma, pois possibilita visualizar a **distribuição** dos preços. Ele é o tipo de visualização mais adequado para esta pergunta porque permite analisar a distribuição dos preços e observar se existem padrões, como uma maior concentração em valores mais baixos ou mais altos.

**7. Existe variação temporal nas vendas?**

Hipótese: As vendas variam ao longo do tempo (tendência ou sazonalidade).

Análise: Agrupamento por order_purchase_timestamp.

A hipótese se confirma parcialmente: observa-se uma clara tendência de crescimento nas vendas ao longo do tempo, acompanhada de variações significativas. No entanto, não há evidência suficiente no gráfico para afirmar a presença de sazonalidade.

```python
vendas_diarias = clean.set_index('order_purchase_timestamp').resample('D')['qtd_item'].count().reset_index()

plt.figure(figsize=(12,6))
plt.plot(vendas_diarias['order_purchase_timestamp'], vendas_diarias['qtd_item'])

plt.title('Vendas Diárias ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Quantidade de Itens Vendidos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Foi utilzado o gráfico de linha e ele é ideal para mostrar a variação de quantidade de itens vendidos ao longo do tempo, isso permite visualizar padrões sazonais .

**8. Pedidos com mais itens têm maior valor total?**

Hipótese:Quanto maior a quantidade de itens, maior o valor total do pedido.

Análise: Relação entre qtd_item e price.

Existe uma relação positiva entre a quantidade de itens e o valor total, porém com alta variabilidade, indicando que pedidos maiores nem sempre resultam em valores mais altos.

```python
pedido = clean.groupby('order_id').agg({
    'qtd_item':'count',
    'price':'sum'
}).reset_index()

plt.figure(figsize=(8,6))
plt.scatter(pedido['qtd_item'], pedido['price'])

plt.title('Itens por Pedido vs Valor Total')
plt.xlabel('Quantidade de Itens')
plt.ylabel('Valor Total')
plt.tight_layout()
plt.show()
```

Aqui foi utilizado o gráfico de dispersão (scatter plot). Nesse caso, foi aplicado para verificar se há relação entre a quantidade de itens e o valor total dos pedidos, permitindo avaliar se pedidos com mais itens tendem a apresentar maior valor total.

**9. As categorias mais vendidas também são as mais bem avaliadas pelos clientes?**

Hipótese: Categorias com maior volume de vendas não são necessariamente as mais bem avaliadas.

Análise: Cruzamento entre product_category_name e review_score médio.

A análise da relação entre volume de vendas e avaliação média, por meio de gráfico de dispersão e correlação, mostrou que não há associação linear significativa entre as variáveis (correlação = -0.0242). Isso indica que categorias mais vendidas não são necessariamente as mais bem avaliadas pelos clientes, confirmando a hipótese inicial.

```python
# Criando o DataFrame category_review agrupando por categoria e calculando o total de itens e a média da avaliação
category_review = clean.groupby('product_category_name').agg(
    total_items=('qtd_item', 'count'),
    avg_review_score=('review_score', 'mean')
).reset_index()

plt.figure(figsize=(8, 6))
plt.scatter(category_review['total_items'], category_review['avg_review_score'], alpha=0.6)

plt.title('Relação entre Volume de Vendas e Avaliação Média')
plt.xlabel('Volume de Vendas (total de itens)')
plt.ylabel('Avaliação Média')
plt.ylim(0, 5)

plt.grid(True)
plt.tight_layout()
plt.show()

# Correlação
correlation = category_review['total_items'].corr(category_review['avg_review_score'])
print(f'Correlação entre volume e avaliação: {correlation:.4f}')
```

Por fim, foi utilizado um gráfico de dispersão (scatter plot), pois esse tipo de visualização é adequado para analisar a relação entre duas variáveis numéricas, o volume de vendas e a avaliação média das categorias. Isso permite observar e identificar padrões ou correlações dos dados.
